```python
# Make plots appear inside the notebook
%matplotlib inline

# Numpy: fast matrix multiplication
import numpy as np

# Matplotlib: plot data nicely
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Scikit-learn: Load Data
from sklearn import datasets

# Scikit-learn: Supervised Learning Models
from sklearn.tree import DecisionTreeClassifier

# Scikit-learn: Evaluation utilities
from sklearn import cross_validation
```

<html>
<body>
<h4>Load data</h4>
</body>
</html>

```python
# load the iris datasets
iris = datasets.load_iris()
attributes = iris.feature_names
for f in attributes:
    print(f)

# Select subset of data
X = iris.data[:, 2:4]
y = iris.target
```

<html>
<body>
<h4>Plot data</h4>
</body>
</html>

```python
# set up plotting
plt.figure(figsize=(10, 8))
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("Anderson's Iris Dataset")

# Plot iris dataset
plt.scatter(X[:,0], X[:,1], c=y, s=160)
plt.show()
```

<html>
<body>
<h4>Learn Decision Tree - Cross Validation </h4>
</body>
</html>

```python
# Evaluate using Cross Validation
myTree = DecisionTreeClassifier()
scores = cross_validation.cross_val_score(myTree, X, y, cv=5)

print("Scores:", scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

<html>
<body>
<h4>Import Packages</h4>
</body>
</html>

```python
# Make plots appear inside the notebook
%matplotlib inline

# Numpy: fast matrix multiplication
import numpy as np

# Matplotlib: plot data nicely
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Scikit-learn: Load Data
from sklearn import datasets

# Scikit-learn: Supervised Learning Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Scikit-learn: Evaluation utilities
from sklearn import cross_validation
```

<html>
<body>
<h4>Load Data</h4>
</body>
</html>

```python
# Load low resolution Handwritten Digits
digits = datasets.load_digits()

# Reshape the images as vectors
n_samples = len(digits.images)
X = digits.images.reshape((n_samples, -1))
y = digits.target
```

<html>
<body>
<h4>Visualize Data</h4>
</body>
</html>

```python
# Visualize first 10 samples
images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:10]):
    plt.subplot(2, 5, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Sample: %i' % label)
```

<html>
<body>
<h4>Train and Evaluate Decision Trees Model</h4>
</body>
</html>

```python
# Create a Decision Tree classifier
myTree = DecisionTreeClassifier()

# Evaluate using Cross Validation
scores = cross_validation.cross_val_score(myTree, X, y, cv=5)

print("Scores:", scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

<html>
<body>
<h4>Train and Evaluate Random Forests</h4>
</body>
</html>

```python
# Create a Random Forest classifier
myForest = RandomForestClassifier(n_estimators=10)

# Evaluate using Cross Validation
scores = cross_validation.cross_val_score(myForest, X, y, cv=5)

print("Scores:", scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
# Make plots appear inside the notebook
%matplotlib inline

# Numpy: fast matrix multiplication
import numpy as np

# Matplotlib: plot data nicely
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data loading and transformation
import pandas as pd

# Scikit-Learn: Regression models
from sklearn import linear_model

# Scikit-Learn: Utilities
from sklearn import datasets
```

<html>
<body>
<h4>Load Dataset</h4>
</body>
</html>

```python
# Load advertisement/Sales Dataset
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
data.head()
```

<html>
<body>
<h4>Visualize Data</h4>
</body>
</html>

```python
# visualize the relationship between the features and the response using scatterplots
fig, axs = plt.subplots(1, 3, sharey=True)
data.plot(kind='scatter', x='TV', y='Sales', ax=axs[0], figsize=(16, 8))
data.plot(kind='scatter', x='Radio', y='Sales', ax=axs[1])
data.plot(kind='scatter', x='Newspaper', y='Sales', ax=axs[2])
```

<html>
<body>
<h4>Univariate Linear Regression</h4>
</body>
</html>

```python
# Shuffle Data
data.reindex(np.random.permutation(data.index))

# Train test split
tv_train = list(data.iloc[range(0,150)].TV.reshape(-1, 1))
sales_train = list(data.iloc[range(0,150)].Sales.reshape(-1, 1))
tv_test = list(data.iloc[range(150,200)].TV.reshape(-1, 1))
sales_test = list(data.iloc[range(150,200)].Sales.reshape(-1, 1))

# Train the model using the training sets
regr = linear_model.LinearRegression()
regr.fit(tv_train, sales_train)

# Evaluate
print("Mean Squared Error (Train): %.2f" % np.mean((regr.predict(tv_train) - sales_train) ** 2))
print("Mean Squared Error (Test): %.2f" % np.mean((regr.predict(tv_test) - sales_test) ** 2))

# Plot Test set
plt.scatter(tv_test, sales_test,  color='black')
plt.plot(tv_test, regr.predict(tv_test), color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```

<html>
<body>
<h4>Non Linear Regression (Ads/Sales)</h4>
</body>
</html>

```python
# Import additional utility functions
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Train and plot non linear regression model
for degree in [2]: #16
    model = make_pipeline(PolynomialFeatures(degree), linear_model.LinearRegression())
    model.fit(tv_train, sales_train)

    x_plot = np.linspace(0, 290, 350).reshape(-1, 1) 
    sales_pred = model.predict(x_plot)
    plt.scatter(tv_train, sales_train,  color='black')
    plt.plot(x_plot, sales_pred, label="degree %d" % degree, linewidth=3)
plt.legend(loc='upper left')
plt.xlabel('TV')
plt.ylabel('Sales')
```

<html>
<body>
<h4>Non Linear Regression (x*sin(x))</h4>
</body>
</html>

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

def f(x):
    return np.sin(x)

# generate points used to plot
x_plot = np.linspace(0, 10, 30).reshape(-1, 1) 

# generate points and keep a subset of them for training
x = np.linspace(0, 10, 30).reshape(-1, 1) 
rng = np.random.RandomState(3)
rng.shuffle(x)
x = np.sort(x[:20])
y = f(x)

# Plot ground function and training points
plt.plot(x_plot, f(x_plot), label="ground truth")
plt.scatter(x, y)

for degree in [4, 6]: #10
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    model.fit(x, y)
    y_plot = model.predict(x_plot)
    plt.plot(x_plot, y_plot, label="degree %d" % degree)
plt.legend(loc='lower left')

plt.show()
```

<html>
<body>
<h4>Non Linear Regression (Overfitting Noise)</h4>
</body>
</html>

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

def f(x):
    return 20*x

def fnoisy(x):
    curr = 20*x + np.random.normal(0, 20, len(x)).reshape(-1, 1) 
    return curr
    
# generate points used to plot
x_plot = np.linspace(0, 10, 30).reshape(-1, 1) 

# generate points and keep a subset of them for training
x = np.linspace(0, 10, 30)
rng = np.random.RandomState(3)
rng.shuffle(x)
x = np.sort(x[:15]).reshape(-1, 1) 
y = fnoisy(x)

# Plot ground function and training points
plt.plot(x_plot, f(x_plot), label="ground truth")
plt.scatter(x, y)

for degree in [1,8]: #10
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    model.fit(x, y)
    y_plot = model.predict(x_plot)
    plt.plot(x_plot, y_plot, label="degree %d" % degree)
plt.legend(loc='lower right')

plt.show()
```
